```python
cuisine_df=df.dropna(subset=["Cuisines"]) #removes missing values from cuisines
```

```python
top_cuisines = cuisine_df['Cuisines'].value_counts().head(3)
```

```python
print("Top 3 Most Common Cuisines:") # task 1-Topmost 3 cuisies
print(top_cuisines)
```

```python
#percentage of top resturants which serves as topmost in cuisines
total_restaurants = len(cuisine_df)

percentage = (top_cuisines / total_restaurants) * 100

print("\nPercentage of Restaurants Serving Top Cuisines:")
print(percentage.round(2))
```

```python
import matplotlib.pyplot as plt  #visulazation
plt.figure(figsize=(8,5))

top_cuisines.plot(kind='bar')

plt.title('Top 3 Most Common Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)

plt.show()
```

```python
results = pd.DataFrame({
    'Count': top_cuisines.values,
    'Percentage (%)': percentage.values
})

results.index = top_cuisines.index

results
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Dataset .csv')
```

```python
print(df.head())
print(df.columns)
```

```python
df.head()
```

```python
city_counts = df['City'].value_counts() #City with the Highest Number of Restaurants
print(city_counts)
```

```python
print("City with Highest Number of Restaurants:")
print(city_counts.head(1))
```

```python
top10 = city_counts.head(10)
plt.figure(figsize=(10,5))
top10.plot(kind='bar')
plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
average_rating = df.groupby('City')['Aggregate rating'].mean()#Average Rating of Each City
print(average_rating)
```

```python
average_rating = average_rating.round(2)#Average Rating of Each City
print(average_rating)
```

```python
average_rating = average_rating.sort_values(ascending=False)#Sort Ratings
```

```python
top10_rating = average_rating.head(10)
plt.figure(figsize=(10,5))
top10_rating.plot(kind='bar')
plt.title("Top 10 Cities by Average Restaurant Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
highest_city = average_rating.idxmax()#Highest Average Rating

highest_rating = average_rating.max()

print("City with Highest Average Rating:", highest_city)

print("Average Rating:", highest_rating)
```

```python
summary = pd.DataFrame({
    "Average Rating": average_rating
})

summary.head(10)
```

**Level 1 Task 3**

```python
price_counts = df['Price range'].value_counts().sort_index()#Count Restaurants in Each Price Range

print("Number of Restaurants in Each Price Range:")
print(price_counts)
```

```python
price_percentage = (price_counts / len(df)) * 100 #Calculate Percentage

print("\nPercentage of Restaurants in Each Price Range:")
print(price_percentage.round(2))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(price_counts.index.astype(str), price_counts.values)

plt.title("Distribution of Restaurant Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(8,5)) #Histogram

plt.hist(df['Price range'], bins=4, edgecolor='black')

plt.title("Histogram of Restaurant Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Frequency")

plt.show()
```

```python
results = pd.DataFrame({    #Results Table
    'Price Range': price_counts.index,
    'Restaurant Count': price_counts.values,
    'Percentage (%)': price_percentage.round(2).values
})

results
```

**Level 1 Task 4**

```python
print(df['Has Online delivery'].value_counts())
```

```python
delivery_percentage = (   #the percentage of restaurants with and without online delivery
    df['Has Online delivery']
    .value_counts(normalize=True) * 100
)

print("Percentage of Restaurants with Online Delivery:")
print(delivery_percentage.round(2))
```

```python
delivery_table = pd.DataFrame({           #table showing the restaurant count and percentage
    'Restaurant Count': df['Has Online delivery'].value_counts(),
    'Percentage (%)': delivery_percentage.round(2)
})

delivery_table
```

```python
average_rating = (  #the average rating for restaurants
# with and without online delivery
    df.groupby('Has Online delivery')['Aggregate rating']
    .mean()
    .round(2)
)

print("Average Ratings:")
print(average_rating)
```

```python
import matplotlib.pyplot as plt #Bar Chart for Online Delivery Percentage

plt.figure(figsize=(6,5))

delivery_percentage.plot(kind='bar')

plt.title("Percentage of Restaurants Offering Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Percentage (%)")
plt.xticks(rotation=0)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(6,5))    #Bar Chart for Average Ratings

average_rating.plot(kind='bar')

plt.title("Average Ratings by Online Delivery")
plt.xlabel("Online Delivery")
plt.ylabel("Average Rating")
plt.xticks(rotation=0)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
